# [dbo].[Vista_OcupacionSalas]

```sql
/****** Object:  View [dbo].[Vista_OcupacionSalas]    Script Date: 19/11/2024 20:07:57 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
CREATE VIEW [dbo].[Vista_OcupacionSalas] AS
SELECT 
    p.NOMBRE AS NombrePelicula,
    s.NOMBRE_SALA AS NombreSala,
    s.CAPACIDAD AS CapacidadSala,
    (SELECT COUNT(t.NUM_TICKET)
     FROM TICKETS t
     JOIN FUNCIONES f ON t.ID_FUNCION = f.ID_FUNCION
     WHERE f.ID_SALA = s.ID_SALA AND f.ID_PELICULA = p.ID_PELICULA) * 100.0 / NULLIF(s.CAPACIDAD, 0) AS PorcentajeOcupacion,
    c.NOM_CLIENTE AS NombreCliente,
    c.APE_CLIENTE AS ApellidoCliente,
    t.FECHA AS FechaCompra
FROM 
    PELICULAS p
JOIN 
    FUNCIONES f ON p.ID_PELICULA = f.ID_PELICULA
JOIN 
    SALAS s ON f.ID_SALA = s.ID_SALA
JOIN 
    TICKETS t ON f.ID_FUNCION = t.ID_FUNCION
JOIN 
    RESERVAS r ON t.ID_RESERVA = r.ID_RESERVA
JOIN 
    CLIENTES c ON r.ID_CLIENTE = c.ID_CLIENTE;
GO
```

# [dbo].[sp_UltimoCliente]

```sql
/****** Object:  StoredProcedure [dbo].[sp_UltimoCliente]    Script Date: 19/11/2024 20:07:57 ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
Create procedure [dbo].[sp_UltimoCliente]
as

select c.NOM_CLIENTE + ' '+ c.APE_CLIENTE 'Nombre del Cliente',
		c.ID_CLIENTE 'Id de Cliente',
		f.FECHA_FACTURA 'Fecha de Factura',
		f.NUM_FACTURA 'Numero de Factura',
		SUM(DF.CANTIDAD*DF.NUM_FACTURA) 'Total Facturado'
from FACTURAS f
join CLIENTES c on f.ID_CLIENTE = c.ID_CLIENTE
join DETALLE_FACTURAS DF ON DF.NUM_FACTURA = F.NUM_FACTURA
where year(f.FECHA_FACTURA) = YEAR(GETDATE())
group by c.NOM_CLIENTE + ' '+ c.APE_CLIENTE, c.ID_CLIENTE, f.NUM_FACTURA, f.FECHA_FACTURA
GO
```
